```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as no
%matplotlib inline
```

```python
#dividing data
X=ds[["Petrol_tax","Average_income","Paved_Highways","Population_Driver_licence(%)"]]
y=ds[["Petrol_Consumption"]]
```

```python
#splitting data into testing and training
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#build the algorithm
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
```

```python
#train the model on the dataset
rg.fit(X_train,y_train)
```

```python
#make predictions
y_pred=rg.predict(X_test)
```

```python
#dataframe-Predicted values vs actual values
df=pd.DataFrame({'Actual Values':y_test.values.flatten(),'Predicted Values':y_pred.flatten()})
display(df)
```

```python
#measure the accuracy of the algorithm
from sklearn import metrics
print("mean absoulute error:", metrics.mean_absolute_error(y_test,y_pred))
```

```python
print("r2 score:",metrics.r2_score(y_test,y_pred))
```

#Fidings
* Mean Absolute error: Tells us the difference between the predicted values and the actual values.


```python
#accuracy of model
print("Accuracy of model:", rg.score(X,y)*100)
```

#fidings
* Linear regression accurately predicted petrol consumption by 67%.
